```python
df_tested_ArunachalPradesh=df_trial[df_trial.state=='Arunachal Pradesh']
#df_tested_ArunachalPradesh_updatedon = df_tested_ArunachalPradesh['updatedon']
#df_tested_ArunachalPradesh_positive = df_tested_ArunachalPradesh['positive']
list_positive_ArunachalPradesh = df_tested_ArunachalPradesh['positive'].to_list()
list_updatedon_ArunachalPradesh = df_tested_ArunachalPradesh['updatedon'].to_list()

list_positive_ArunachalPradesh = [int(i) for i in list_positive_ArunachalPradesh]


df_tested_ArunachalPradesh
```

```python
df_tested_HimachalPradesh=df_trial[df_trial.state=='Himachal Pradesh']
#df_tested_HimachalPradesh_updatedon = df_tested_HimachalPradesh['updatedon']
#df_tested_HimachalPradesh_positive = df_tested_HimachalPradesh['positive']

list_positive_HimachalPradesh = df_tested_HimachalPradesh['positive'].to_list()
list_updatedon_HimachalPradesh = df_tested_HimachalPradesh['updatedon'].to_list()

list_positive_HimachalPradesh = [int(i) for i in list_positive_HimachalPradesh]


df_tested_HimachalPradesh
```

```python
df_tested_Tripura=df_trial[df_trial.state=='Tripura']

#df_tested_Tripura.sort_values(by='updatedon')


#df_tested_Tripura_updatedon = df_tested_Tripura['updatedon']
#df_tested_Tripura_positive = df_tested_Tripura['positive']

list_positive_Tripura = df_tested_Tripura['positive'].to_list()
list_updatedon_Tripura = df_tested_Tripura['updatedon'].to_list()

list_positive_Tripura = [int(i) for i in list_positive_Tripura]


df_tested_Tripura
```

```python
from datetime import datetime
plt.figure(figsize=(18,10))

date_objects_Goa = [datetime.strptime(date, '%d/%m/%Y').date() for date in list_updatedon_Goa]
#date_objects_Karnataka = [datetime.strptime(date, '%d/%m/%Y').date() for date in list_updatedon_Karnataka]
date_objects_ArunachalPradesh = [datetime.strptime(date, '%d/%m/%Y').date() for date in list_updatedon_ArunachalPradesh]
date_objects_HimachalPradesh = [datetime.strptime(date, '%d/%m/%Y').date() for date in list_updatedon_HimachalPradesh]
date_objects_Tripura = [datetime.strptime(date, '%d/%m/%Y').date() for date in list_updatedon_Tripura]


plt.plot(date_objects_Goa, list_positive_Goa, marker='+',color='red', label = "Goa")
#plt.plot(date_objects_Karnataka, list_positive_Karnataka, marker='>',color='orange', label = "Karnataka")
plt.plot(date_objects_ArunachalPradesh, list_positive_ArunachalPradesh,marker='x',color='blue', label = "Arunachal Pradesh")
plt.plot(date_objects_HimachalPradesh, list_positive_HimachalPradesh,marker='o',color='green', label = "Himachal Pradesh")
plt.plot(date_objects_Tripura, list_positive_Tripura,marker='^',color='orange', label = "Tripura")

plt.title('Comparision of positive cases for states having similar population as of Goa ')

plt.legend()

plt.show()
```
